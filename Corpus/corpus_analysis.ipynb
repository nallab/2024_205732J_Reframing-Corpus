```python
# 2.年齢
import re
import matplotlib.pyplot as plt

ages = []
for i in range(len(persona_df)):
  age = persona_df["年齢"][i]
  age = re.sub(r"\D", "", age)
  ages.append(int(age))

under_teens = []
teens = []
twenties = []
thirties = []
forties = []
fifties = []
sixties_and_over = []
for age in ages:
  if age < 10:
    under_teens.append(age)
  if 10 <= age < 20:
    teens.append(age)
  if 20 <= age < 30:
    twenties.append(age)
  if 30 <= age < 40:
    thirties.append(age)
  if 40 <= age < 50:
    forties.append(age)
  if 50 <= age < 60:
    fifties.append(age)
  if 60 <= age:
    sixties_and_over.append(age)

x_age = [1, 2, 3, 4, 5, 6, 7]
y_count = [len(under_teens), len(teens), len(twenties), len(thirties), len(forties), len(fifties), len(sixties_and_over)]
labels = ["~9", "10's", "20's", "30's", "40's", "50's", "60~"]

plt.bar(x_age, y_count, width=0.5, tick_label=labels)

for i in range(len(x_age)):
    plt.text(x_age[i], y_count[i] + 0.05, str(y_count[i]), ha='center')

plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

# 発話内容分析

```python
sheet_name = "発話"
corpus_df = pd.read_excel(excel_file_path, sheet_name=sheet_name)
```

```python
corpus_df.head()
```

```python
# 各発話の長さ
corpus_df["発話長"] = corpus_df["発話"].str.len()

# 最大長
max_length = corpus_df["発話長"].max()
print("最大長の文字数:", max_length)

# 最小長
min_length = corpus_df["発話長"].min()
print("最小長の文字数:", min_length)
```

```python
# ポジティブ返答(奇数インデックス)の行
odd_index_df = corpus_df.iloc[1::2]

# "発話長"の最小値を持つ行のインデックス
min_length_index = odd_index_df["発話長"].idxmin()

min_length_index
```

```python
# ネガティブ発話（偶数インデックス）とポジティブ返答（奇数インデックス）をペアとして組み合わせ
paired_statements = pd.DataFrame({
    "ネガティブ発話": corpus_df.iloc[::2].reset_index(drop=True)["発話"],
    "ポジティブ返答": corpus_df.iloc[1::2].reset_index(drop=True)["発話"]
})

# ユニークなペアの数をカウント
unique_pairs_count = paired_statements.drop_duplicates().shape[0]
unique_pairs_count
```

```python
# ネガティブ発話のユニーク数
negative_statements = corpus_df.iloc[::2].reset_index(drop=True)["発話"]
unique_negative_statements = negative_statements.nunique()
print("ネガティブ発話のユニーク数:", unique_negative_statements)
```

```python
# ポジティブ返答のユニーク数
positive_statements = corpus_df.iloc[1::2].reset_index(drop=True)["発話"]
unique_positive_statements = positive_statements.nunique()
print("ポジティブ返答のユニーク数:", unique_positive_statements)
```

```python
# ポジティブ返答で5回以上出現するもの
positive_statements = corpus_df.iloc[1::2].reset_index(drop=True)["発話"]
positive_statements_count = positive_statements.value_counts()
positive_statements_count.head(10)
positive_statements_count[positive_statements_count >= 5]
```

```python
# ポジティブ返答の出現頻度の度数分布
occurrences = list(range(1, 25))
counts = [496, 63, 16, 15, 3, 8, 4, 4, 1, 1, 2, 0, 2, 1, 0, 0, 0, 0, 0, 0, 2, 0, 0, 1]

plt.figure(figsize=(12, 8))

plt.yscale('log')
bars = plt.bar(occurrences, counts)

for bar in bars:
    yval = bar.get_height()
    if yval > 0:
        plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')

plt.xlabel('Number of Occurrences')
plt.ylabel('Count')
plt.xticks(occurrences)
plt.grid(axis='y', which='both', linestyle='')

plt.tight_layout()
plt.show()
```

```python
import MeCab

mecab = MeCab.Tagger()

# 形態素解析をして単語数をカウントする
def count_words(text):
    node = mecab.parseToNode(text)
    count = 0
    while node:
        if node.surface:
            count += 1
        node = node.next
    return count

# 新しい列「単語数」を追加
corpus_df["単語数"] = corpus_df["発話"].apply(count_words)

corpus_df.head()
```

```python
# ネガティブ発話の平均単語数
negative_statements_words = corpus_df.iloc[::2].reset_index(drop=True)["単語数"]
negative_statements_words_mean = negative_statements_words.mean()
print("ネガティブ文の平均単語数:", negative_statements_words_mean)

# ポジティブ返答の平均単語数
positive_statements_words = corpus_df.iloc[1::2].reset_index(drop=True)["単語数"]
positive_statements_words_mean = positive_statements_words.mean()
print("ポジティブ文の平均単語数:", positive_statements_words_mean)
```

```python
# 文節数をカウントする関数
def count_phrases(text):
    mecab.parse('')
    node = mecab.parseToNode(text)
    phrases_count = 0
    while node:
        # 文節の始まりを示す品詞をカウント
        if node.feature.startswith('名詞') or node.feature.startswith('動詞') or node.feature.startswith('形容詞') or node.feature.startswith('副詞'):
            phrases_count += 1
        node = node.next
    return phrases_count

# 新たな列「文節数」を追加
corpus_df["文節数"] = corpus_df["発話"].apply(count_phrases)

corpus_df.head()
```

```python
# ネガティブ発話の平均文節数
negative_statements_clause = corpus_df.iloc[::2].reset_index(drop=True)["文節数"]
negative_statements_clause_mean = negative_statements_clause.mean()
print("ネガティブ発話の平均文節数:", negative_statements_clause_mean)

# ポジティブ返答の平均文節数
positive_statements_clause = corpus_df.iloc[1::2].reset_index(drop=True)["文節数"]
positive_statements_clause_mean = positive_statements_clause.mean()
print("ポジティブ返答の平均文節数:", positive_statements_clause_mean)
```
